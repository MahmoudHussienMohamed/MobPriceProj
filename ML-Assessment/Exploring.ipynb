# Data-set first look

```python
%matplotlib inline
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
```

```python
DSDIR = 'train.csv'
LABEL = 'price_range'
```

Let's take a look at first 10 instances from data:

```python
ds = pd.read_csv(DSDIR)
ds.head()
```

And see the data types of each column and whether any of them have null values or not:

```python
ds.info()
```

And some statistics:

```python
ds.describe()
```

Correlation matrix to know the most important features:

```python
corr_matrix = ds.corr()
price_corr = corr_matrix[LABEL].sort_values(ascending=False)
price_corr
```

```python
def plotCorrelationMatrix(df, graphWidth):
    figname = 'Mobile Price'
    corr = df.corr()
    plt.figure(num=None, figsize=(graphWidth, graphWidth), dpi=80, facecolor='w', edgecolor='k')
    corrMat = plt.matshow(corr, fignum = 1)
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
    plt.yticks(range(len(corr.columns)), corr.columns)
    plt.gca().xaxis.tick_bottom()
    plt.colorbar(corrMat)
    plt.title(f'Correlation Matrix for {figname}', fontsize=15)
    plt.show()
```

To plot the values that has the strongest correlation:

```python
features = price_corr.index
threshold = 0.02
highest_corr_features = [
    feature for feature in features 
    if price_corr[feature] >= threshold or price_corr[feature] <= -threshold
]
print(highest_corr_features)
plotCorrelationMatrix(ds[highest_corr_features], 20)
```
